```python
#I am going to perform one-hot-encoding to make sure we can use all the columns to the fullest
def YorN(data):
    if data == 'Yes': return 1
    else: return 0
```

```python
#utilizng the funtion we created on all the Columns that have Yes and No as entries
farmer_data['Aadhar'] = farmer_data['Aadhar Card'].apply(YorN)
farmer_data.drop('Aadhar Card',axis=1,inplace=True)
```

```python
farmer_data['Farmer Portal'] = farmer_data['Registered on Farmer Portal'].apply(YorN)
farmer_data.drop('Registered on Farmer Portal',axis=1,inplace=True)
```

```python
farmer_data['Gov_Scheme'] = farmer_data['Government Scheme Beneficiary'].apply(YorN)
farmer_data.drop('Government Scheme Beneficiary',axis=1,inplace=True)
```

```python
farmer_data['Aadhar'] = farmer_data['Advanced Agriculture Equipment'].apply(YorN)
farmer_data.drop('Advanced Agriculture Equipment',axis=1,inplace=True)
```

```python
farmer_data['Owners'] = farmer_data['Land Owners'].apply(YorN)
farmer_data.drop('Land Owners',axis=1,inplace=True)
```

```python
#Lets take alook at the reformed farmer_data
farmer_data.head()
```

```python
#Now time to handle all particular isntance of one-hot-encoding.
from collections import defaultdict
def encoder(name):
    dic = defaultdict(list)
    count=0
    for uniq in farmer_data[name].unique():
        dic[uniq].append(count)
        count+=1
    def solver(a,dic=dic):
        return dic[a][0]
    naming_system = name.split()
    new_name = '-'.join(naming_system)
    farmer_data[new_name] = farmer_data[name].apply(solver)
```

### Data Conversion into type required for ML Models

```python
#Lets Drop the name column, as it contributes nothin to the target variable
#Splitting the dataset into Features(X) and Target(Y)
X_farm = farmer_data.drop('Loan_Granted',axis=1)
X_farm.drop('names of farmers',axis=1,inplace=True)
Y_farm = farmer_data['Loan_Granted']
```

```python
# Splitting the dataset into traning and testing sets
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_farm,Y_farm)
```

```python
#Let's utilize the Principal Component Analysis(PCA) to select only essential features, which will make our model faster, and will provide us with a decreased loss.
from sklearn.decomposition import PCA
```

```python
pca = PCA()
x_train_dummy_pca = pca.fit_transform(x_train)
x_test_dummy_pca = pca.transform(x_test)
```

```python
pca.explained_variance_
```

```python
#to find and take only the best possible componenets for optimal results
j = 0
current_variance = 0
total = sum(pca.explained_variance_)
while current_variance/total < 0.90:
    current_variance += pca.explained_variance_[j]
    j+=1
j
```

```python
#Despite Decreasing the accracy level to 90%, the pca suggested components are not raising more than 2, however, using such
#low amount of features to train our model might result in overfitting, and might not be very scalable to the real-world.
#So it makes more sense to ignore the PCA, and pass all our current features to the model.
```

```python
y_train = y_train.ravel() #to change shape of  Y_test/Y_train into desired shape
y_train
```

```python
y_test = y_test.ravel()
y_test
```

## Model Creation

```python
#Importing required models
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline #To create the pipeline.
from sklearn.model_selection import GridSearchCV #To hyper tune the parameters
from sklearn.preprocessing import StandardScaler
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto', probability=True),
        'params': {
            'svc__C': [1, 10, 100, 1000],
            'svc__kernel': ['rbf', 'linear']
        }
    },
    'random_forest_classifier': {
        'model': RandomForestClassifier(),
        'params': {
            'randomforestclassifier__n_estimators': [1, 5, 10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'logisticregression__C': [1, 5, 10]
        }
    }
}
```

```python
#To calculate the best scores, along with the best_predictors for every different model in our pipeline
scores = []
best_estimators = {}
for algo,mp in model_params.items():
    pipe = make_pipeline(StandardScaler(),mp['model'])
    clf = GridSearchCV(pipe,mp['params'],cv=5,return_train_score=False)
    clf.fit(x_train,y_train)
    scores.append({
        'model':algo,
        'best_score': clf.best_score_,
        'best_params':  clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
```

```python
data = pd.DataFrame(scores,columns=['model','best_score','best_params'])
data
```

```python
best_estimators
```

```python
best_estimators['random_forest_classifier'].score(x_test,y_test)
```

```python
best_predictor = best_estimators['random_forest_classifier']
```

```python
from sklearn.metrics import confusion_matrix
confuse = confusion_matrix(y_test,best_predictor.predict(x_test))
confuse #creating the confusion matrix of the y_test and predicted_x_test
```
