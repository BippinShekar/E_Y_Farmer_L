```python
#renaming the column to names of farmers
farmers.rename(columns={'names of farmer': 'names of farmers'},inplace=True)
```

#### Making necessary additions to the databse - Data Creation

```python
## Creating a column that will indicate if the farmer is registered with the government or not, i.e the farmer has a Aadhar Card or not.
## As the Live Aadhar, as in the person holding the Aadhar who are live is 127.79 crores out of total pouplation of India of 140.76 crores.
#It is safe to assume almost 90% of farmers have aadhar cards
per90 = int((len(farmers)*90)/100)
def aadharY(a):
    return 'Yes'
farmers['Aadhar Card'] = farmers['names of farmers'][:per90].apply(aadharY)
```

```python
#The remaining 10% people/farmers that don't have aadhar cards
def aadharN(a):
    return 'No'
farmers['Aadhar Card'][per90:] = farmers['names of farmers'][per90:].apply(aadharN)
```

```python
#Shuffling the total Aadhar Card column, to make sure it is random, that which farmer has addhar and which doesn't.
random.shuffle(farmers['Aadhar Card'])
```

```python
# Creating a column that will indicatate if the farmer hass been registered to the Farmer's Portal offered by the Indian Government
def creator(a):
    return random.choice(('Yes','No'))
farmers['Registered on Farmer Portal'] = farmers['names of farmers'].apply(creator)
```

```python
## Creating a column that indicates if the farmer has been apporved for any Governement Schemes
## Pradhan Mantri Fasal Bima Yojana (PMFBY) - Agricultiural Insurance
## Soil Health Card 
##Pradhan Mantri Krishi Sinchai Yojana (PMKSY)
##National Agriculture Market (e-NAM)
## PARAMPARAGAT KRISHI VIKAS YOJNA (PKVY) 
def Govscheme(a):
    return random.choice(('Yes','No'))
farmers['Government Scheme Beneficiary'] = farmers['names of farmers'].apply(Govscheme)
```

```python
## creating a column that will signify the Land Area Type of the Land owned by the farmer
def landt(a):
    return random.choice(('Marginal','Small','Semi-Medium','Medium','Large'))
farmers['Land Class'] = farmers['names of farmers'].apply(landt)
```

```python
#Creating a column that will store, the total area of the Land owned by the farmer, based on the category/class under which the farm falls under by size.
# The land area is in hectare, 1ha = 2.471 acres
def landaread(a):
    if a=='Marginal': return random.choice((0.80,0.85,0.90,0.95,0.98))
    elif a=='Small': return random.choice((1,1.08,1.5,1.8,2))
    elif a=='Semi-Medium': return random.choice((2,2.5,2.9,3,3.5,3.9))
    elif a=='Medium': return random.choice((4,5,6,7,8,9))
    else: return random.choice((10,11,13,15,17,20))
farmers['Land Area'] = farmers['Land Class'].apply(landaread)
```

```python
## Accoring to Prime Minister of India, around 85% of farmers own lands, so we are creating a Land Owner column, that is split in 80 Owners and 20 Non owners, because we are assuming a deviation of 5% to make sure generalization is being done.
def owner80(a):
    return 'Yes'
per80 = int((len(farmers)*80)/100) 
farmers['Land Owners'] = farmers['names of farmers'][:per80].apply(owner80)
```

```python
def notowner20(a):
    return 'No'
farmers['Land Owners'][per80:] = farmers['names of farmers'][per80:].apply(notowner20)
```

```python
print(farmers['Land Owners'].head(),farmers['Land Owners'].tail())
```

```python
## Shuffling the land owners column, to make sure we are distributing the probability of farmers being owners and not, in the required ration of 80-20
random.shuffle(farmers['Land Owners'])
```

```python
## creating a column that will hold the current land Value, of the farmers land.
## avg value of farming land is 0.60 crores per acre
def landval(a):
    return random.choice(((a*2.471)*0.6,(a*2.471)*0.55))
farmers['Land Value - Crs'] = farmers['Land Area'].apply(landval)
```

```python
#Creating  a column that will indicate how is the health of the soil of the land that the farmer is cultivating on.
def soil(a):
    return random.choice(('Good','Medium','Low'))
farmers['Soil Health'] = farmers['names of farmers'].apply(soil)
```

```python
## creating a column that holds all the different types of crops the farmers in india can grow
def croptypes(a):
    return random.choice(('PaddyC','PaddyA','JowarH','JowarM','Bajra','Ragi','Maize','Arhar','Moong','Urad','CottonM','CottonL','Groundnut','Sunflower','Soyabean','Sesamum','Nigersseed','Wheat','Barley','Gram','Lentil','Mustard','Safflower','Sugarcane'))
farmers['Crop Type'] = farmers['names of farmers'].apply(croptypes)
```

```python
## Creating a crop cost column, this will hold the cost that is required to grow a quintol of the crop chosen by the farmer for his piece of land.
def cropcost(a):
    if a == 'PaddyC': return 1360
    elif a == 'PaddyA': return 1360
    elif a == 'JowarH': return 1977
    elif a=='JowarM': return 1977
    elif a == 'Bajra':return 1268
    elif a=='Ragi': return 2385
    elif a=='Maize': return 1308
    elif a=='Arhar': return 4131
    elif a=='Moong': return 5167
    elif a=='Urad': return 4155
    elif a=='CottonM': return 4053
    elif a=='CottonL': return 4053
    elif a=='Groundnut': return 3873
    elif a=='Sunflower': return 4113
    elif a=='Soyabean': return 2805
    elif a=='Sesamum': return 5220
    elif a=='Nigerseed': return 4858
    elif a=='Wheat': return 1065
    elif a=='Barley': return 1082
    elif a=='Gram': return 3206
    elif a=='Lentil': return 3239
    elif a=='Mustard': return 2670
    elif a=='Safflower': return 3765
    elif a == 'Sugarcane': return 157
farmers['Crop Cost - Qtl'] = farmers['Crop Type'].apply(cropcost)
```

```python
## To create a coulmn called Type Of Farming, to make sure we understand which niche of farming the farmers belong to
def niche(a):
    return random.choice(('Organic','Commercial','Subsistence','Intensive'))
farmers['Type of Farming'] = farmers['names of farmers'].apply(niche)
```

```python
## This column will be used to hold, if the farmer tha'ts asking for laon has help to the farm his land or not.
def employee(a):
    return random.choice(('Family','Yes','No'))
farmers['Employes'] = farmers['names of farmers'].apply(employee)
```

```python
## Creating a important factor of Irrigation and its Type, which is essential for crop survival and harvest.
def irrigator(a):
    return random.choice(('Manual Irrigation','Sprinkler Irrigation','Surface Irrigation'))
farmers['Irrigation Mode'] = farmers['names of farmers'].apply(irrigator)
```

```python
# Here by Advanced Agri Equip we mean, tools like power tiller, or brush cutter, Thresher.
def aaequip(a):
    return random.choice(('Yes','No'))
farmers['Advanced Agriculture Equipment'] = farmers['names of farmers'].apply(aaequip)
```

```python
## TO create a proper mode of harvesting column, to know and understand the type of harvesting the partciular farmer uses, to estimate his loan repayablitiy
def modeh(a):
    return random.choice(('Manual','Hand-Held Tools','Machinery'))
farmers['Mode of Harvesting'] = farmers['names of farmers'].apply(modeh)
```

```python
## CWC - Central warehousing corporation - Central Govt intiative to build storages at select places to hold crop harvests.
## SWC - State warrehousing corporation - State Govt + CWC at centers of disctrict importance.
## FCI - Food corporation of India - has a capacity of 26.62 million tons.
## Gramin Bhandaran - Scheme to provide place to store harvest by helping rural godowns.
## Cold Storage - Less in quantity, however, good for long term storage.
## Open Land - The farmers that don't have access to any facilities mentioned above, will store the harvest in the land itself, by putting them in bags and covering them by a big plastic cover.
def storage(a):
    return random.choice(('CWC','ColdStorage','Gramin Bhandaran','SWC','Open Land','FCI'))
farmers['Storage Types'] = farmers['names of farmers'].apply(storage)
```

```python
# Creating the column, which will hold the money at which on succesull harvesting the farmer can seel the total harvest for. This MSP Column shos the minimum sale price set by the Indian Governement.
def msp(a):
    if a == 'PaddyC': return 2040
    elif a == 'PaddyA': return 2060
    elif a == 'JowarH': return 2970
    elif a=='JowarM': return 2990
    elif a == 'Bajra':return 2350
    elif a=='Ragi': return 3578
    elif a=='Maize': return 1962
    elif a=='Arhar': return 6600
    elif a=='Moong': return 7755
    elif a=='Urad': return 6600
    elif a=='CottonM': return 6080
    elif a=='CottonL': return 6380
    elif a=='Groundnut': return 5850
    elif a=='Sunflower': return 6400
    elif a=='Soyabean': return 4300
    elif a=='Sesamum': return 7830
    elif a=='Nigerseed': return 7287
    elif a=='Wheat': return 2125
    elif a=='Barley': return 1735
    elif a=='Gram': return 5335
    elif a=='Lentil': return 6000
    elif a=='Mustard': return 5450
    elif a=='Safflower': return 5650
    elif a == 'Sugarcane': return 315
farmers['MSP-Qunitol'] = farmers['Crop Type'].apply(msp)
```

```python
## Creating a column that will hold, how many qunitols can the particular farmers land produce.
#Lets assume a ha, produces around 55-60 qunitols of yield.
def possyeild(a):
    return random.choice((a*55,a*60))
farmers['Total Possible Yeild'] = farmers['Land Area'].apply(possyeild)
```

```python
#Creating a column that will hold the total possible profit based on the gathered information until now.
def totalcropcost(row):
    total_cost = row['Crop Cost - Qtl'] * row['Total Possible Yeild']
    return total_cost
farmers['Total Crop Cost'] = farmers.apply(totalcropcost,axis=1)
```

```python
#Creating a column that will hold total possible revenue, bassed on total possible yeild produced by the farmer.
# Lets assume 10% of the farmers total yeild has been wasted, the rest 90% of his yeild will account to his total revenue
def possrev(row):
    rev_yield = row['Total Possible Yeild'] - ((row['Total Possible Yeild'] * 10) / 100)
    total_rev = rev_yield * row['MSP-Qunitol']
    return total_rev

farmers['Total Revenue'] = farmers.apply(possrev, axis=1)
```

```python
#A column to calculate the total possible profit, ie Total revenue - total cost
def totprofit(row):
    total_prof = row['Total Revenue'] - row['Total Crop Cost']
    return total_prof
farmers['Total Profit'] = farmers.apply(totprofit,axis=1)
```

```python
## THis column indicates the loan amount being asked by the farmer
def reqloan(a):
    return random.choice((a+((a*10)/100),a+((a*20)/100),a+((a*25)/100),a+((a*30)/100)))
farmers['Loan Request'] = farmers['Total Crop Cost'][:per80].apply(reqloan)
```

```python
def reqloanup(a):
    return random.choice((a+((a*40)/100),a+((a*50)/100)))
farmers['Loan Request'][per80:] = farmers['Total Crop Cost'][per80:].apply(reqloanup)
```

```python
## Loan Granted - The Target Variable
## This final column will hold, if the farmer has been granted the loan or not, by considerng all major factors.
def granted(row):
    if row['Aadhar Card'] == 'No' or row['Land Owners'] == 'No':
        return 'No'
    else:
        if row['Loan Request'] > (row['Total Crop Cost'] + ((row['Total Crop Cost']*40)/100)): return 'No'
        else: 
            if row['Government Scheme Beneficiary'] == 'No' and row['Soil Health'] == 'low': return 'No'
            else:
                if row['Storage Types'] == 'Open Land' and row['Government Scheme Beneficiary']=='No': return 'No'
                else:
                    if row['Land Class'] == 'Medium' and row['Employes'] == 'No': return 'No'
                    else: return 'Yes'
    return 'Yes'
farmers['Loan Granted'] = farmers.apply(granted,axis=1)
```
